```python
df_populacao = pd.read_excel('POP2022_Municipios.xls', header=1, skipfooter=34)
df_populacao
```

```python
df_populacao.info()
```

```python
import seaborn as sns
```

```python
#df_populacao['POPULAÇÃO'] = df_populacao['POPULAÇÃO'].astype('int64')
valid_numbers = pd.to_numeric(df_populacao['POPULAÇÃO'], errors='coerce')
invalid_numbers = valid_numbers.isna()
df_populacao[invalid_numbers]
```

```python
df_populacao.loc[invalid_numbers, 'POPULAÇÃO'] = df_populacao.loc[invalid_numbers, 'POPULAÇÃO'].str.replace(r'(\.|\(\d+\))', '', regex=True)
df_populacao[invalid_numbers]
```

```python
df_populacao['POPULAÇÃO'] = df_populacao['POPULAÇÃO'].astype(int)
```

```python
df_populacao
```

```python
df_populacao.dtypes
```

```python
df_populacao_uf = df_populacao.groupby('UF')[['POPULAÇÃO']].sum().reset_index()
df_populacao_uf
```

```python
df_emissao_uf = df_ano[df_ano['Ano'] == 2021].groupby('Estado')[['Emissão']].sum().reset_index()
df_emissao_uf
```

```python
df_agrupados = pd.merge(df_emissao_uf, df_populacao_uf, left_on='Estado', right_on='UF', how='left')
df_agrupados
```

```python
df_agrupados.plot(x='POPULAÇÃO', y='Emissão', kind='scatter', figsize=(8,6))
```

```python
sns.scatterplot(df_agrupados, x='POPULAÇÃO', y='Emissão', hue='Estado', palette='Accent')
```

```python
df_agrupados = df_agrupados.assign(emissao_per_capita=df_agrupados['Emissão']/df_agrupados['POPULAÇÃO']).sort_values('emissao_per_capita', ascending=False)
df_agrupados
```

```python
sns.barplot(
    df_agrupados,
    x='Estado',
    y='emissao_per_capita',
    palette='Accent')
```

```python
sns.scatterplot(
    df_agrupados,
    x='POPULAÇÃO',
    y='Emissão',
    hue='Estado',
    size='emissao_per_capita',
    palette='Accent')
```
